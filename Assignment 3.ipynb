## Method
I use a multiple regression model where the [GINI index](https://www.investopedia.com/terms/g/gini-index.asp) is measure of income inequality per country *i* per year *t* - a Gini coefficient of zero expresses perfect equality, where everyone has the same income. My variable of interest is $UMP_t$: this is a dummy variable that assumes value 1 if in a particular year the ECB implemented an unconventional monetary policy and 0 otherwise. Hence, I first run the following regression, including a first and second lag of the dependent variable to solve the likely presence of autocorrelation: 

$$ Gini_{i,t} = \alpha + \beta_0 UMP_t + \beta_1 Gini_{i,t-1} + \beta_2 Gini_{i,t-2} + \varepsilon_t .$$

In the [Robustness](## Robustness) section, I will further extend my model.

## Unconventional policies 

When a nation's economy grows to the point that inflation increases to dangerous levels, the central bank will pursue restrictive monetary policy to tighten the money supply and reduce the amount of money in circulation. When a nation's economy slips into recession, this policy tool can be operated in reverse: interest rates are lowered, reserve limits loosened and bonds are purchased in exchange for newly created money.

The issue with such conventional policies is that in periods of deep recession they become limited. According to [RBI’s Deepak Mohanty](https://www.bis.org/review/r131227b.pdf), "When central banks look beyond their traditional instrument of policy interest rate, monetary
policy takes an unconventional character." In my analysis I will focus on four unconventional monetary policies implemented by the ECB as a response to the last financial crisis: [SMP](https://www.ecb.europa.eu/pub/pdf/other/mb201006_focus01.en.pdf?2b5c64d483a41a019fd461f595a36b46), [LTROs](https://www.ecb.europa.eu/press/pr/date/2011/html/pr111208_1.en.html), [OMT](https://www.ecb.europa.eu/press/pr/date/2012/html/pr120906_1.en.html) and [QE](https://www.ecb.europa.eu/explainers/show-me/html/app_infographic.en.html).


#### Securities Markets Programme (SMP)

The [“Securities Markets Programme” (SMP)](https://www.ecb.europa.eu/pub/pdf/other/mb201006_focus01.en.pdf?2b5c64d483a41a019fd461f595a36b46), announced on 10 May 2010, is intended to ensure depth and liquidity in malfunctioning segments of the debt
securities markets and to restore an appropriate functioning of the monetary policy transmission
mechanism; the objective is to restore the effective conduct of monetary policy oriented towards price stability in the medium term.
The impact of these interventions is sterilised offering commercial banks the possibility to invest central bank money at the ECB as a time deposit: through such specific operations, the liquidity injected is re-absorbed and thereby the monetary policy stance is not affected. The SMP simply involved the purchase of bonds – especially sovereign bonds – on the secondary markets. Owing to the ban on monetary financing, no purchases were made on the primary market, and the last purchase was made in February 2012. At its peak, the programme's volume totalled around €210 billion. 



Source: [*blogs.reuters.com*](http://blogs.reuters.com/globalinvesting/2012/04/16/three-snapshots-for-monday-5/)



#### Outright Monetary Transactions (OMT)
On 2 August 2012, the Governing Council of the European Central Bank (ECB) [announced](http://www.ecb.europa.eu/press/pr/date/2012/html/pr120906_1.en.html) that it would undertake outright transactions in secondary, sovereign bond markets, aimed at safeguarding an appropriate monetary policy transmission and the singleness of the monetary policy. The technical framework of these operations was formulated on 6 September 2012. On the same date, the bank's [SMP](### Securities Markets Programme (SMP) was terminated.
For the OMT to be activated towards a certain eurozone state, a total of four conditions needs to be fully met:
1. **Conditionality:**
The state needs to have received financial sovereign support from the eurozone's bailout funds [EFSF](https://www.esm.europa.eu/investors/efsf/funding-strategy)/[ESM](https://www.esm.europa.eu/investors/esm/funding-strategy).
1. **Coverage:**
No ex ante quantitative limits are set on the size of Outright Monetary Transactions.
1. **Creditor treatment:**
The Eurosystem clarifies in the legal act concerning Outright Monetary Transactions that it accepts the same (pari passu) treatment as private or other creditors with respect to bonds issued by euro area countries and purchased by the Eurosystem through Outright Monetary Transactions, in accordance with the terms of such bonds.
1. **Sterilization:**
The liquidity created through Outright Monetary Transactions will be fully sterilised.

The OMT has never been implemented. Its only announcement, by the way, has had [positive effects in reducing bond yields](http://uk.businessinsider.com/ecj-decision-on-omt-legal-2015-1?IR=T).




Source: [*BusinessInsider.com*](http://www.businessinsider.com/mario-draghi-greece-contagion-2015-8?IR=T)


#### Longer-Term Refinancing Operations (LTROs)

[LTROs](https://www.ecb.europa.eu/press/pr/date/2011/html/pr111208_1.en.html)  provide an injection of low interest rate funding to euro zone banks with sovereign debt as collateral on the loans. Until early 2008, the longest LTRO maturity was three months. Since then the ECB has successively introduced six-month, 12-month and 36-month terms for LTRO finance. Each of these new issues has been heavily subscribed, with eurozone periphery banks in Ireland, Italy, Spain and Greece taking the majority of the first 36-month issue in late 2011. The second 36-month issue in February 2012 was closely watched as an indicator of the health of eurozone banks. Since early 2008, the loans are offered monthly and are typically repaid in three months, six months or one year. The second 36-month issue in February 2012 was closely watched as an indicator of the health of eurozone banks. The goal of such policy is twofold: encouraging euro zone banks to increase lending activities that spur the economy by providing access to cheap capital and lowering sovereign debt [yields](https://www.thebalance.com/what-is-the-yield-curve-416919) since countries can use their own sovereign bonds as collateral, with a consequent increase in their demand. 


Source: [*NCBs*](https://www.ecb.europa.eu/euro/changeover/estonia/html/location.en.html), [*Goldman Sachs Research*](http://www.goldmansachs.com/what-we-do/research/)



#### Quantitative Easing (QE)
The ECB started buying assets from commercial banks in March 2015 as part of its non-standard monetary policy measures. These asset purchases, also known as [QE](https://www.ecb.europa.eu/explainers/show-me/html/app_infographic.en.html), support economic growth across the euro area and help to return to inflation levels below, but close to, 2%.
By buying eurozone governments’ bonds using electronically created money, the ECB hoped to shore up confidence, boost inflation, and drive down the value of the euro on foreign exchanges, helping to boost exports and kickstart growth. The €60bn a month used includes the purchases of private sector assets to try to unlock credit markets. Mario Draghi [announced](https://www.theguardian.com/business/live/2015/jan/22/davos-2015-day-2-qe-ecb-lagarde-draghi-merkel-live?page=with:block-54c1131ee4b0990b136292fb#block-54c1131ee4b0990b136292fb) that he would pump €1.1tn at a rate of €60bn (£46bn) a month into financial markets until September 2016, in an attempt to prevent the fragile eurozone economy from grinding to a halt.

## Robustness

To check the robustness of my [results](## Results) I try to extend my model. The findings of [Jaumotte et al. (2008)](https://www.imf.org/external/pubs/ft/wp/2008/wp08185.pdf) suggest that the most significant covariates for
explaining change in income inequality are the credit to the private sector, the
ratio of inward FDI (Foreign Direct Investment) to GDP and the industrial
employment share. The first two variables have a positive correlation with
income inequality: that is, an increase in one of the two causes an increase
in Gini, while industrial employment share is negatively associated with Gini.
Furthermore, the export-to-GDP ratio is significantly negatively correlated with Gini, which suggests that trade
globalization is associated with lower income inequality (ibid.). Thus I run now the following regression, taking into account inflation:
$$ Gini_{i,t} = \alpha + \beta_0 UMP_t + \beta_1 Gini_{i,t-1} + \beta_2 Gini_{i,t-2} + \beta_3 \frac{Credit_{i,t}}{GDP_{i,t}} + \beta_4 \frac{FDI_{i,t}}{GDP_{i,t}} + \beta_5 Industrial_{{i,t}} + \beta_6 \frac{Export_{i,t}}{GDP_{i,t}} + \beta_7 \Pi_{i,t} + \varepsilon_t .$$

```python
# Creating dependent and independent variables for my multiple regression
robtarget = pd.DataFrame(filtered_data,columns=["GINI"])
robpredictors=pd.DataFrame(filtered_data,columns=['UMP','GINIt-1','GINIt-2','Credit_GDP','GDP_growth','Inflation','FDI_GDP'])
# Regression
mult_regression = LinearRegression()
mult_regression.fit(robpredictors,robtarget)
```

```python
# Coefficients
mult_regression.coef_
```

```python
# Intercept
mult_regression.intercept_
```

#### Results
Here the results are quite different. The estimated regression is:
$$ Gini_{i,t} = 1.20396309 + 0.02584784 UMP_t + 0.87556172 Gini_{i,t-1} + 0.07286237 Gini_{i,t-2} +  0.00209014 \frac{Credit_{i,t}}{GDP_{i,t}} + 0.00099131 \frac{FDI_{i,t}}{GDP_{i,t}} - 0.05836862 Industrial_{{i,t}} + 0.02873237 \frac{Export_{i,t}}{GDP_{i,t}} + 0.00437564 \Pi_{i,t}.$$

From what we see, the presence of unconventional monetary policy raised inequality. The coefficient of my variable of interest - $UMP_t$ - is indeed positive.

## Discussion

The opposite results may be due to the fact that in the second model I included more variables which are high-effective predictors of income inequality. Here, the results make more sense if we consider how unconventional policies work: they simply provide the creation of wealth through injection of money in the markets or in the banking system. Usually, the users of these channels already own a high level of wealth, and as a result they become even richer if compared to the poor. Of course, over the years, such increase of richness will be redistributed to the lower income-classes for example through higher consumptions, but this requires time. To date, maybe the policies are so recent that, in the second and probably more reliable regression I have run, we can not observe positive redistributive effects.
